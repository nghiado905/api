```python
# Hàm Random_alpha sinh các hàm bọc xung quanh
group = ['sector', 'industry', 'subindustry', 'market', 'exchange', 'market', 'country']
days = [5,10,30,50,80,100,120,200,250]

def Random_alpha(expression, dodai_alpha, nhom_thuong, nhom_ts, nhom_group, nhom_2bien):
    if dodai_alpha <= 0:
        return expression
    else:
        i = random.randrange(1, 10)
        m = random.randint(1, 5)
        n = random.randint(4, 9)
        if i in [1, 2, 9]:
            return nhom_group[random.randrange(len(nhom_group))] + \
                   '(' + Random_alpha(expression, dodai_alpha - (m or n), nhom_thuong, nhom_ts, nhom_group, nhom_2bien) + \
                   ',' + group[random.randrange(len(group))] + ')'
        elif i in [3, 4, 5]:
            return nhom_ts[random.randrange(len(nhom_ts))] + \
                   '(' + Random_alpha(expression, dodai_alpha - (m or n), nhom_thuong, nhom_ts, nhom_group, nhom_2bien) + \
                   ',' + str(random.choice(days)) + ')'
        elif i in [6, 7, 8]:
            return nhom_thuong[random.randrange(len(nhom_thuong))] + \
                   '(' + Random_alpha(expression, dodai_alpha - m, nhom_thuong, nhom_ts, nhom_group, nhom_2bien) + ')'
        else:  # i == 10
            return nhom_2bien[random.randrange(len(nhom_2bien))] + \
                   '(' + Random_alpha(expression, dodai_alpha - 1, nhom_thuong, nhom_ts, nhom_group, nhom_2bien) + \
                   ',' + Random_alpha(expression, dodai_alpha - 1, nhom_thuong, nhom_ts, nhom_group, nhom_2bien) + \
                   ',' + str(random.randint(1, 30)) + ')'
```
